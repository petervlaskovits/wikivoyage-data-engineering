```python
from scrapers import WikipediaScraper, WikivoyageScraper
import pandas as pd
import time
```

```python
wikipedia = WikipediaScraper()
wikivoyage = WikivoyageScraper()
```

```python
wikipedia_df = wikipedia.get_world_cities()
wikipedia_df.to_csv("data/global_cities_list.csv", index=False)
```

```python
# generate parquet file
cities = wikipedia_df.city
wikivoyage_df = pd.DataFrame()

total_cities = len(cities)
counter = 0

for city in cities:
    print(f"Attempting to retrieve city {city}")
    city_df = wikivoyage.get_by_car_row(city)
    print(city_df.section_text)
    wikivoyage_df = pd.concat([wikivoyage_df, city_df])
    counter += 1
    print(f"Retrieved city {city}, progress: {counter} / {total_cities}")
    time.sleep(3)

print("Finished scraping loop!")
```

```python
df = pd.merge(left=wikivoyage_df, right=wikipedia_df, on="city")
df.to_parquet("data/city_wikivoyage_articles_uncleaned.parquet")
```
