```python
df = df.drop(labels=[234, 235, 161, 162, 362, 363])
```

```python
duplicated_cities = df.duplicated(subset=["city"])
df[duplicated_cities == True]
```

```python
df["subregion"] = df.wiki_article.str.split(",_").str[1]
```

```python
df
```

```python
filtered_df = df[(df["article_status"] == "Unknown/Stub")]
filtered_df
```

```python
wv = WikivoyageScraper()

def replace_row(article_name, replaced_index):
    row = wv.get_by_car_row(article_name)
    other_row = filtered_df.loc[replaced_index]

    row.at[0, "country"] = other_row.country
    row.at[0, "wiki_article"] = other_row.wiki_article
    row.at[0, "subregion"] = other_row.subregion

    ndf = pd.concat([filtered_df, row]).drop(labels=[replaced_index])
    return ndf
    
fixed_cities = [
    ("New York City", 1),
    ("Delhi", 39), 
    ("San Jose (Costa Rica)", 159),
    ("San Jose (California)", 160),
    ("Córdoba (city, Argentina)", 246),
    ("Grand Cayman", 258),
    ("Hamilton (Bermuda)", 261),
    ("Łódź", 278),
    ("Portland (Oregon)", 303),
    ("San Luis Potosi", 313),
    ("Aguascalientes", 340),
    ("Kingston (Jamaica)", 352),
    ("León (Mexico)", 354),
    ("Mérida (Mexico)", 356),
    ("Rochester (New York)", 396)
]

for fixed_city in fixed_cities:
    filtered_df = replace_row(fixed_city[0], fixed_city[1])
    time.sleep(5)
    
filtered_df = filtered_df.reset_index(drop=True)
```

```python
filtered_df
```

```python
old_indexes = [x[1] for x in fixed_cities]

df = df.drop(index=old_indexes)
df = pd.concat([df, filtered_df]).reset_index(drop=True)
```
